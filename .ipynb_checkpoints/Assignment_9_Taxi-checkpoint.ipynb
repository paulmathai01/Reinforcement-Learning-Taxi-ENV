```python
import gym
import numpy as np
import time
```

```python
alpha = 0.4
gamma = 0.999
epsilon = 0.9
episodes = 10000
max_steps = 2500
n_tests = 1
render = False
test = True
env = gym.make('Taxi-v3')
n_states, n_actions = env.observation_space.n, env.action_space.n
Q = np.ones((n_states, n_actions))
timestep_reward = []
for episode in range(episodes):
    s = env.reset()
    if np.random.rand() < epsilon:
        a = np.argmax(Q[s, :])
    else:
        a = np.random.randint(0, n_actions)
    t = 0
    total_reward = 0
    done = False
    while t < max_steps:
        if render:
            env.render()
        t += 1
        s_, reward, done, info = env.step(a)
        total_reward += reward
        a_ = np.argmax(Q[s_, :])
        if done:
            Q[s, a] += alpha * ( reward  - Q[s, a] )
        else:
            Q[s, a] += alpha * ( reward + (gamma * Q[s_, a_]) - Q[s, a] )
        s, a = s_, a_
        if done:
            if render:
                print(f"This episode took {t} timesteps and reward: {total_reward}")
            timestep_reward.append(total_reward)
            break
for test in range(n_tests):
    print(f"Test #{test+1}")
    s = env.reset()
    done = False
    epsilon = 0
    while True:
        time.sleep(1)
        env.render()
        a = np.argmax(Q[s, :])
        print(f"Chose action {a} for state {s}")
        s, reward, done, info = env.step(a)
        if done:
            if reward > 0:
                print("Reached goal!")
            else:
                print("Failed")
            time.sleep(3)
            break
```

```python
plt.figure(figsize=(10,10))
plt.plot(timestep_reward)
plt.ylabel('Reward')
plt.legend(["Qlearning"])
plt.show()
```
